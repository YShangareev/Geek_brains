```python
import numpy as np
import pandas as pd
import sklearn as skl
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.datasets import fetch_california_housing
from sklearn.ensemble import RandomForestRegressor
```

```python
data = fetch_california_housing()
X = pd.DataFrame(data=data.data, columns=data.feature_names)
y = data.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 4)
```

```python
std_scale = StandardScaler()
X_train_scale = pd.DataFrame(std_scale.fit_transform(X_train), columns=X_train.columns)
X_test_scale = pd.DataFrame(std_scale.transform(X_test), columns=X_test.columns)
```

```python
lr = LinearRegression()
lr.fit(X_train_scale, y_train)
pred = lr.predict(X_test_scale)

r2_test = lr.score(X_test_scale, y_test)
r2_train = lr.score(X_train_scale, y_train)
print(f'Метрика (R2)\nна обучении:{round(r2_train,2)}\nна тесте:{round(r2_test,2)}')
```

```python
df = pd.read_csv('/content/creditcard.csv')
```

```python
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, roc_curve
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
pd.options.display.max_columns = 100
```

```python
df['Class'].value_counts(normalize=True)
```

```python
df.info()
```

```python
df.head(10)
```

```python
X = df.iloc[:,:df.shape[1]-1]
y = df.iloc[:,df.shape[1]-1:]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3, random_state=10, stratify=y)
```

```python
parametrs = [{'n_estimators':[10,15],'max_features':np.arange(3,5),'max_depth':np.arange(4,7)}]
clf = GridSearchCV(estimator=RandomForestClassifier(random_state=4), param_grid=parametrs, scoring='roc_auc',cv=3)
clf.fit(X_train,y_train)
clf.best_params_
```

```python
y_pred_proba_test = clf.predict_proba(X_test)[:,1]
y_pred_proba_train = clf.predict_proba(X_train)[:,1]
```

```python
R_AUC_test = roc_auc_score(y_test, y_pred_proba_test)
R_AUC_train = roc_auc_score(y_train, y_pred_proba_train)
print(f'Метрика (ROC_AUC)\nОбучение:{round(R_AUC_test,2)}\nТест:{round(R_AUC_train,2)}')
```

```python
plt.figure(figsize=(16,16))
fpr_test,tpr_test,thresh_test = roc_curve(y_test, y_pred_proba_test,pos_label=1)
fpr_train,tpr_train,thresh_train = roc_curve(y_train, y_pred_proba_train,pos_label=1)
plt.plot(fpr_test,tpr_test)
plt.plot(fpr_train,tpr_train)
plt.legend(['test', 'train'])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC curve')
```
